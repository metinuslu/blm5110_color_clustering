```python
# Create Dataset with Images

# Renkler
colors = ['red', 'blue', 'green', 'gray', 'white']
# colors = ['sample']

# Boş bir DataFrame oluştur
df_data = pd.DataFrame(columns=['FilePath', 'FileName', 'Color'])

for color in colors:
    color_path = os.path.join(DATA_PATH, 'train', color)

    # Renk altındaki dosyaları al
    files = os.listdir(color_path)

    # DataFrame'e eklemek için veriyi oluştur
    data = {'FilePath':color_path, 'FileName': files, 'Color': color,}

    # DataFrame'e ekle
    df_data = pd.concat([df_data, pd.DataFrame(data)], ignore_index=True)
```

```python
def create_img_arr(file_names):
    image_arr_list = []
    width_list = []
    height_list = []
    channel_list = []
    
    for file in file_names:
        img = cv2.imread(file)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV, renkleri BGR formatında okur, bu nedenle RGB'ye dönüştürüyoruz

        # Resmi yeniden boyutlandır
        # new_width = 600
        # new_height = 800      
        # image = cv2.resize(image, (new_width, new_height))        
        
        width, height, channel = img.shape
        
        # Görseli düzleştirerek vektöre dönüştür
        # img = img.reshape((-1, 3))

        width_list.append(width)
        height_list.append(height)
        channel_list.append(channel)
        
        # image_arr_list.append(img)
        image_arr_list.extend(img)
        
        # Tüm matrisleri bir NumPy dizisine dönüştür
        # image_arr = np.array(image_arr_list)
        
    return image_arr_list, width_list, height_list, channel_list
```

```python
def kmeans(X, k, max_iters=100):
    """
    K-Means algoritması implementasyonu.

    Parametreler:
    - X: Veri seti (numpy array)
    - k: Küme sayısı
    - max_iters: Maksimum iterasyon sayısı

    Döndürülen değerler:
    - centroids: Hesaplanan küme merkezleri
    - labels: Her örneğin ait olduğu küme etiketleri
    """

    # Veri setindeki örnek sayısı
    m = X.shape[0]

    # Veri setinden rastgele k örnek seçilerek küme merkezleri başlatılır
    centroids = X[np.random.choice(m, k, replace=False)]

    for _ in range(max_iters):
        # Her örnek için en yakın küme merkezini bulma
        labels = np.argmin(np.linalg.norm(X[:, np.newaxis] - centroids, axis=2), axis=1)

        # Yeni küme merkezlerini hesaplama
        new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(k)])

        # Küme merkezlerinin değişimini kontrol etme
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return centroids, labels
```
